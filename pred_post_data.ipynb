```python
import pandas as pd
```

```python
import xml
import xml.dom
import os
from os import listdir
```

```python
path=r'C:\Users\Annie\Documents\Working\Корпус биографических текстов\corpus\\'
f_list=os.listdir(path)
```

```python
sentences=[]
types=[]
pred=[]
predpred=[]
predpredpred=[]
pred4=[]
post=[]
postpost=[]
postpostpost=[]
post4=[]

ii=0

for f in f_list:
    try:
        doc=xml.dom.minidom.parse(path+f)
        items=doc.getElementsByTagName('sentence')

        new_items=[]
        for i, elem in enumerate(items):
            type_fact=elem.getElementsByTagName('type')
            type_fact=type_fact[0].firstChild.data
            if str(type_fact)!='other' and str(type_fact)!='none':
                new_items.append(items[i])

        for i, elem in enumerate(new_items):

            num=i+1

            type_fact=elem.getElementsByTagName('type')
            types.append(type_fact[0].firstChild.data.replace('-1','birth').replace('nationality','parenting'))

            if str(num)=='1':
                pred.append('begin')
                predpred.append('begin')
                predpredpred.append('begin')
                pred4.append('begin')
            elif str(num)=='2':
                pred.append(new_items[i-1].getElementsByTagName('text')[0].firstChild.data)
                predpred.append('begin')
                predpredpred.append('begin')
                pred4.append('begin')
            elif str(num)=='3':
                pred.append(new_items[i-1].getElementsByTagName('text')[0].firstChild.data)
                predpred.append(new_items[i-2].getElementsByTagName('text')[0].firstChild.data)
                predpredpred.append('begin')
                pred4.append('begin')
            elif str(num)=='4':
                pred.append(new_items[i-1].getElementsByTagName('text')[0].firstChild.data)
                predpred.append(new_items[i-2].getElementsByTagName('text')[0].firstChild.data)
                predpredpred.append(new_items[i-3].getElementsByTagName('text')[0].firstChild.data)
                pred4.append('begin')
            else:
                pred.append(new_items[i-1].getElementsByTagName('text')[0].firstChild.data)
                predpred.append(new_items[i-2].getElementsByTagName('text')[0].firstChild.data)
                predpredpred.append(new_items[i-3].getElementsByTagName('text')[0].firstChild.data)
                pred4.append(new_items[i-4].getElementsByTagName('text')[0].firstChild.data)

            if str(num)==str(len(new_items)):
                post.append('end')
                postpost.append('end')
                postpostpost.append('end')
                post4.append('end')
            elif str(num)==str(len(new_items)-1):
                post.append(new_items[i+1].getElementsByTagName('text')[0].firstChild.data)
                postpost.append('end')
                postpostpost.append('end')
                post4.append('end')
            elif str(num)==str(len(new_items)-2):
                post.append(new_items[i+1].getElementsByTagName('text')[0].firstChild.data)
                postpost.append(new_items[i+2].getElementsByTagName('text')[0].firstChild.data) 
                postpostpost.append('end')
                post4.append('end')
            elif str(num)==str(len(new_items)-3):
                post.append(new_items[i+1].getElementsByTagName('text')[0].firstChild.data)
                postpost.append(new_items[i+2].getElementsByTagName('text')[0].firstChild.data) 
                postpostpost.append(new_items[i+3].getElementsByTagName('text')[0].firstChild.data)
                post4.append('end')
            else:
                post.append(new_items[i+1].getElementsByTagName('text')[0].firstChild.data)
                postpost.append(new_items[i+2].getElementsByTagName('text')[0].firstChild.data) 
                postpostpost.append(new_items[i+3].getElementsByTagName('text')[0].firstChild.data)
                post4.append(new_items[i+4].getElementsByTagName('text')[0].firstChild.data)

            text=elem.getElementsByTagName('text')
            sentences.append(text[0].firstChild.data)
    except:
        print(f)
    ii+=1
    if ii==190:
        df=pd.DataFrame(sentences)
        df.to_csv('sentences_train.csv', sep=';')
        df=pd.DataFrame(types)
        df.to_csv('types_train.csv', sep=';')
        df=pd.DataFrame(post)
        df.to_csv('post_train.csv', sep=';')
        df=pd.DataFrame(postpost)
        df.to_csv('postpost_train.csv', sep=';')
        df=pd.DataFrame(postpostpost)
        df.to_csv('postpostpost_train.csv', sep=';')
        df=pd.DataFrame(post4)
        df.to_csv('post4_train.csv', sep=';')
        df=pd.DataFrame(pred)
        df.to_csv('pred_train.csv', sep=';')
        df=pd.DataFrame(predpred)
        df.to_csv('predpred_train.csv', sep=';')
        df=pd.DataFrame(predpredpred)
        df.to_csv('predpredpred_train.csv', sep=';')
        df=pd.DataFrame(pred4)
        df.to_csv('pred4_train.csv', sep=';')
        sentences=[]
        types=[]
        pred=[]
        predpred=[]
        predpredpred=[]
        pred4=[]
        post=[]
        postpost=[]
        postpostpost=[]
        post4=[]
    if ii==195:
        df=pd.DataFrame(sentences)
        df.to_csv('sentences_val.csv', sep=';')
        df=pd.DataFrame(types)
        df.to_csv('types_val.csv', sep=';')
        df=pd.DataFrame(post)
        df.to_csv('post_val.csv', sep=';')
        df=pd.DataFrame(postpost)
        df.to_csv('postpost_val.csv', sep=';')
        df=pd.DataFrame(postpostpost)
        df.to_csv('postpostpost_val.csv', sep=';')
        df=pd.DataFrame(post4)
        df.to_csv('post4_val.csv', sep=';')
        df=pd.DataFrame(pred)
        df.to_csv('pred_val.csv', sep=';')
        df=pd.DataFrame(predpred)
        df.to_csv('predpred_val.csv', sep=';')
        df=pd.DataFrame(predpredpred)
        df.to_csv('predpredpred_val.csv', sep=';')
        df=pd.DataFrame(pred4)
        df.to_csv('pred4_val.csv', sep=';')
        sentences=[]
        types=[]
        pred=[]
        predpred=[]
        predpredpred=[]
        pred4=[]
        post=[]
        postpost=[]
        postpostpost=[]
        post4=[]
   
df=pd.DataFrame(sentences)
df.to_csv('sentences_test.csv', sep=';')
df=pd.DataFrame(types)
df.to_csv('types_test.csv', sep=';')
df=pd.DataFrame(post)
df.to_csv('post_test.csv', sep=';')
df=pd.DataFrame(postpost)
df.to_csv('postpost_test.csv', sep=';')
df=pd.DataFrame(postpostpost)
df.to_csv('postpostpost_test.csv', sep=';')
df=pd.DataFrame(post4)
df.to_csv('post4_test.csv', sep=';')
df=pd.DataFrame(pred)
df.to_csv('pred_test.csv', sep=';')
df=pd.DataFrame(predpred)
df.to_csv('predpred_test.csv', sep=';')
df=pd.DataFrame(predpredpred)
df.to_csv('predpredpred_test.csv', sep=';')
df=pd.DataFrame(pred4)
df.to_csv('pred4_test.csv', sep=';')
```
